```python
#Important: In geoWithin function, if you use longitude and latitude, specify longitude first.
#funcion que recibe un poligono y una coleccion de Mongo y te devuelve todos los puntos de la linea que aquel dia pasaron por aquel poligono
def points_inside_polygon(mongo_collection, inverted_polygon):
    points_inside_polygon = []
    for row in mongo_collection.find ({"latlon": {"$geoWithin": { "$polygon": inverted_polygon }}}).sort("timestamp",pymongo.ASCENDING):
        points_inside_polygon.insert(len(points_inside_polygon), [row[u'timestamp'], row[u'latitude'], row[u'longitude'], row[u'line'], row[u'bus_id']]) 
    df_points_inside_polygon = pd.DataFrame(points_inside_polygon, columns=['timestamp', 'latitude', 'longitude', 'line', 'bus_id'])
    return df_points_inside_polygon
```

```python
"""Funciones para renderizar el mapa en el html"""
from IPython.display import HTML
 
def inline_map(map):
    """
    Embeds the HTML source of the map directly into the IPython notebook.
    
    This method will not work if the map depends on any files (json data). Also this uses
    the HTML5 srcdoc attribute, which may not be supported in all browsers.
    """
    map._build_map()
    return HTML('<iframe srcdoc="{srcdoc}" style="width: 100%; height: 510px; border: none"></iframe>'.format(srcdoc=map.HTML.replace('"', '&quot;')))
 
def embed_map(map, path="map.html"):
    """
    Embeds a linked iframe to the map into the IPython notebook.
    
    Note: this method will not capture the source of the map into the notebook.
    This method should work for all maps (as long as they use relative urls).
    """
    map.create_map(path=path)
    return HTML('<iframe src="files/{path}" style="width: 100%; height: 510px; border: none"></iframe>'.format(path=path))
```

```python
#Funcion que crea la tabla de viajes con los tiempos de viaje
def segment_travel_table (name_A, arrive_array_A, name_B, departure_array_B, bus_Id, bus_line):
    
    my_time_table = []   
    num_travel = 0    
    i = 0 #Cursor para moverme por el departure_array_A(arreglo q contiene las horas de salida del punto incial)   
    
   
    while (i<len(departure_array_B)-1):
        if (arrive_array_A[i] <= departure_array_B[i]):
            direction = name_A + "_to_" + name_B
            departure_hour = departure_array_B[i]
            arrive_hour = arrive_array_A[i]
            travel_time = bus.days_hours_minutes(departure_hour - arrive_hour) 
            i = i +1
            my_time_table.insert(len(my_time_table), [bus_line, bus_Id, direction, arrive_hour, departure_hour, travel_time])
                        
    
    #Construir el dataframe pasandole el array con todos los datos de los viajes   
    mydf = pd.DataFrame(my_time_table, columns=['Line', 'Bus_Id','Segment', 'Arrive_Hour', 'Departure_Hour', 'Travel_Segment_Time'])       
    return mydf 
```

```python
#Funcion que retorna un dataframe con los tiempos que demora cada busId en atravesar el segmento entre el punto_inicio y punto_fin
def segment_travel_time(punto_inicio, punto_fin, date):
    import bus
    new_df = pd.DataFrame(columns=['Line', 'Bus_Id','Segment', 'Arrive_Hour', 'Departure_Hour', 'Travel_Segment_Time']) 
    if (punto_inicio == "punto3") and (punto_fin == "punto6"):
        inverted_polygon = inverted_polygon_3_6
        #bus_line_list = ['110']
        bus_line_list = ['110','111','162','172','177','460','476','570']
    if (punto_inicio == "punto1") and (punto_fin == "punto4"):
        inverted_polygon = inverted_polygon_1_4
        bus_line_list = ['172','177','178','309','316','317','550','565']
    if (punto_inicio == "punto2") and (punto_fin == "punto5"):
        inverted_polygon = inverted_polygon_2_5
        bus_line_list = ['170','172','173','178','309','316','317','410']
    
    if (date == '27-07-2015'):
        conn_string = db.dados20150727
        #conn_string = db.dados20150727_line_110
    if (date == '03-08-2015'):
        conn_string = db.dados20150803
    if (date == '10-08-2015'):
        conn_string = db.dados20150810
    if (date == '17-08-2015'):
        conn_string = db.dados20150817    
    #Retorna todos las observaciones de todos las lineas que pasaron por el segmento aquel dia
    df_points_in_segment = points_inside_polygon(conn_string, inverted_polygon)
    
    #despuÃ©s que ya se determinaron los viajes en el sentido deseado
    for line in bus_line_list:
        #Recuperando las observaciones dentro del segmento por lÃ­nea
        df_buslines_inside_polygon = df_points_in_segment[(df_points_in_segment['line']== line)]
        
        df_buslines_inside_polygon.groupby('line').bus_id.nunique()
        bus_id_list = pd.unique(df_buslines_inside_polygon.bus_id.ravel())
        myentry = []
        myout = []
        for bus_id in bus_id_list:
            df_busId_inside_polygon = df_points_in_segment[(df_points_in_segment['bus_id']== bus_id)]
            timestamp_busId_inside_polygon = df_busId_inside_polygon ['timestamp']
            timestamp_busId_inside_polygon.index = range(0,len(timestamp_busId_inside_polygon))
            myentry = bus.bus_arrive_hour(timestamp_busId_inside_polygon)
            myout = bus.bus_departure_hour(timestamp_busId_inside_polygon)
            result = segment_travel_table(punto_inicio, myentry, punto_fin, myout, bus_id, line)
            new_df = new_df.append(result, ignore_index = True) 
            
    return new_df

      
```

```python
#Funcion para convertir la columna Travel_Segment_Time en entero para poder graficarla
def convert_to_int(column):
    list_time = []
    for row in column:
        split_time_array = row.split(':')
        t_time = int(split_time_array[1])
        list_time.insert(len(list_time),t_time)
    return list_time   
```

```python
#Funcion para separar la hora de la fecha en la columna Arrive_Hour
def convert_to_hour(column):
    from datetime import datetime
    list_time = []
    for row in column:
        str_row = str(row)
        split_time_array = str_row.split(' ')
        t_time = datetime.strptime(split_time_array[1], '%H:%M:%S').time()
        list_time.insert(len(list_time),t_time)
    return list_time
```

```python
#esta funcion se modifico mas abajo
#Funcion que dadas las observaciones de un busId dentro de un poligono ordenadas por timestamp,  
#retorna la lista de viajes dentro de ese poligono para ese determinado busId(tanto en una direcciÃ³n como en la otra)
def travels_inside_polygon(points_in_polygon):
    import datetime
    start_travel_vector = []
    #reindexar el dataframe para si no tiene el indice en orden, pues ponerselo
    points_in_polygon.index = range(0,len(points_in_polygon))
    
    for i in range(0, len(points_in_polygon)-1):
        resto = datetime.datetime.strptime(points_in_polygon['timestamp'][i+1], "%d-%m-%Y %H:%M:%S")-datetime.datetime.strptime(points_in_polygon['timestamp'][i], "%d-%m-%Y %H:%M:%S")
        resto_minutes = resto.total_seconds()/60
        #si la diferencia entre los timestamp del omnibus en el putno supera los 40 minutos es q dio un viaje y regreso, sino lo supera es que se mantuvo ese tiempo esperando en el punto
        if (resto_minutes > 40):
            start_travel_vector.insert(len(start_travel_vector),points_in_polygon['timestamp'][i])
    start_travel_vector.insert(len(start_travel_vector),points_in_polygon['timestamp'][len(points_in_polygon)-1])
    
    travel_list = []
    for i in range(0, len(start_travel_vector)):
        if (i==0):
            temp = points_in_polygon[(points_in_polygon['timestamp']<= start_travel_vector[0])]
            travel_list.insert(len(travel_list),temp)
        else:
            temp = points_in_polygon[(points_in_polygon['timestamp']> start_travel_vector[i-1])& (points_in_polygon['timestamp']<= start_travel_vector[i])]
            travel_list.insert(len(travel_list),temp)
        
    return travel_list
```

```python
def travels_indices(points_in_polygon):
    import datetime
    start_travel_vector = []
    #reindexar el dataframe para si no tiene el indice en orden, pues ponerselo
    points_in_polygon.index = range(0,len(points_in_polygon))
    
    for i in range(0, len(points_in_polygon)-1):
        resto = datetime.datetime.strptime(points_in_polygon['timestamp'][i+1], "%d-%m-%Y %H:%M:%S")-datetime.datetime.strptime(points_in_polygon['timestamp'][i], "%d-%m-%Y %H:%M:%S")
        resto_minutes = resto.total_seconds()/60
        #si la diferencia entre los timestamp del omnibus en el putno supera los 40 minutos es q dio un viaje y regreso, sino lo supera es que se mantuvo ese tiempo esperando en el punto
        if (resto_minutes > 40):
            start_travel_vector.insert(len(start_travel_vector),i)
    start_travel_vector.insert(len(start_travel_vector),len(points_in_polygon)-1)
    return start_travel_vector
```

```python
"""
#FunciÃ³n que dado un punto dentro de un poligono y un vector con posiciones latitud, longitud, timestamp dentro de ese 
poligono de un solo bus_Id
#retorna dataframe con los puntos que pertnecen a los viajes que van en una direccion alejandose del punto
#para el caso de las calles con doble sentido
"""
def bus_direction(point, df_inside_polygon):
    from scipy.spatial import distance
    
    #new_df1 = pd.DataFrame(columns=[ 'timestamp', 'latlon', 'line', 'bus_id'])
    vector_dir = []
          
    #identificar viajes (en ambos sentidos)
    #travels_list = travels_inside_polygon(df_inside_polygon)
    
    #creando una nueva columna al dataframe "latlon" que es la union de las columnas latitude y longitude
    #df_inside_polygon['latlon'] = zip(df_inside_polygon.latitude, df_inside_polygon.longitude)
   
    #determinando indices
    misindices = travels_indices(df_inside_polygon)
        
    #creando el vector con los viajes(conjunto de observaciones) en la direccion deseada
    k = 0
    for j in range(0, len(misindices)):
        for i in range(k, misindices[j]+1):
            if (i+2 < misindices[j]+1):
                if(distance.euclidean(df_inside_polygon['latlon'][i],point) < distance.euclidean(df_inside_polygon['latlon'][i+1],point)) & (distance.euclidean(df_inside_polygon['latlon'][i+1],point) < distance.euclidean(df_inside_polygon['latlon'][i+2],point)):
                    vector_dir.insert (len(vector_dir),(i,df_inside_polygon['latlon'][i]))
                 
            elif (i == misindices[j]-1):
                if(distance.euclidean(df_inside_polygon['latlon'][i],point) < distance.euclidean(df_inside_polygon['latlon'][i+1],point)):
                    vector_dir.insert (len(vector_dir),(i,df_inside_polygon['latlon'][i]))
            else:
                if (distance.euclidean(df_inside_polygon['latlon'][i],point) > distance.euclidean(df_inside_polygon['latlon'][i-1],point)):
                    vector_dir.insert (len(vector_dir),(i,df_inside_polygon['latlon'][i]))
                   
        k = i+1
    return vector_dir 
```

```python
#funcion que te calcula la distancia de todos los puntos dentro del poligono con respecto a un punto P
def new_column_distance_to_point(df_inside_polygon,point):
    from scipy.spatial import distance
    dist_vector = []
    
    #creando una nueva columna al dataframe "latlon" que es la union de las columnas latitude y longitude
    df_inside_polygon['latlon'] = zip(df_inside_polygon.latitude, df_inside_polygon.longitude)
    
    for i in range(0,len(df_inside_polygon)):
        dist_vector.insert(len(dist_vector),distance.euclidean(df_inside_polygon['latlon'][i],point))

    #si la columna 'distance' no existe, creala e inserta en ella los valores q estÃ¡n en el arreglo dist_vector
    if not 'distance_to_point' in df_inside_polygon:
        newName = 'distance_to_point'
        values = np.insert(df_inside_polygon.values,df_inside_polygon.shape[1],dist_vector,axis=1)
        header = df_inside_polygon.columns.values.tolist()
        header.append(newName)
        df_inside_polygon = pd.DataFrame(values,columns=header)
    
    return df_inside_polygon
```

```python
#funcion que dado un viaje me dice si va en la direccion opuesta de un punto P
def direc(travel_inside_polygon):
    vector_dir = []
    indexes = list(travel_inside_polygon.index)
    
    #primer indice de travel_inside_polygon
    indexes[0]
    #Ãºltimo indice de travel_inside_polygon
    indexes[len(indexes)-1]
    
    #reindexando travel_inside_polygon 
    travel_inside_polygon.index = range(0,len(travel_inside_polygon))
    
    for i in range(0, len(travel_inside_polygon)):
        
        if (i+2 < len(travel_inside_polygon)):
            if(travel_inside_polygon['distance_to_point'][i] < travel_inside_polygon['distance_to_point'][i+1]) & (travel_inside_polygon['distance_to_point'][i+1] < travel_inside_polygon['distance_to_point'][i+2]):
                vector_dir.insert (len(vector_dir),[indexes[i],travel_inside_polygon['timestamp'][i],travel_inside_polygon['latlon'][i],travel_inside_polygon['line'][i],travel_inside_polygon['bus_id'][i],travel_inside_polygon['distance_to_point'][i]])
        
        elif (i == len(travel_inside_polygon)-2):
            if(travel_inside_polygon['distance_to_point'][i] < travel_inside_polygon['distance_to_point'][i+1]):
                vector_dir.insert (len(vector_dir),[indexes[i],travel_inside_polygon['timestamp'][i],travel_inside_polygon['latlon'][i],travel_inside_polygon['line'][i],travel_inside_polygon['bus_id'][i],travel_inside_polygon['distance_to_point'][i]])
        
        elif(i == len(travel_inside_polygon)-1):
            if (travel_inside_polygon['distance_to_point'][i] > travel_inside_polygon['distance_to_point'][i-1]):
                vector_dir.insert (len(vector_dir),[indexes[i],travel_inside_polygon['timestamp'][i],travel_inside_polygon['latlon'][i],travel_inside_polygon['line'][i],travel_inside_polygon['bus_id'][i],travel_inside_polygon['distance_to_point'][i]])
        
    df_travel = pd.DataFrame(vector_dir, columns=['travel_index', 'timestamp', 'latlon', 'line', 'bus_id','distance_to_point'])
    
    return  df_travel  

```
