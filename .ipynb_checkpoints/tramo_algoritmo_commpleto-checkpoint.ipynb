```python
#Important: In geoWithin function, if you use longitude and latitude, specify longitude first.
#funcion que recibe un poligono y una coleccion de Mongo y te devuelve todos los puntos de la linea que aquel dia pasaron por aquel poligono
def points_inside_polygon(mongo_collection, inverted_polygon):
    points_inside_polygon = []
    for row in mongo_collection.find ({"latlon": {"$geoWithin": { "$polygon": inverted_polygon }}}).sort("timestamp",pymongo.ASCENDING):
        points_inside_polygon.insert(len(points_inside_polygon), [row[u'timestamp'], row[u'latitude'], row[u'longitude'], row[u'line'], row[u'bus_id']]) 
    df_points_inside_polygon = pd.DataFrame(points_inside_polygon, columns=['timestamp', 'latitude', 'longitude', 'line', 'bus_id'])
    return df_points_inside_polygon
```

```python
#Funcion que recibe un dataframe con todas las observaciones de un único bus_id de un line que están dentro de un poligono
#retorna una lista con los indices de donde termina cada viaje de ese bus_id dentro del poligono
def travels_indices(points_in_polygon):
    import datetime
    start_travel_vector = []
    #reindexar el dataframe para si no tiene el indice en orden, pues ponerselo
    points_in_polygon.index = range(0,len(points_in_polygon))
    
    for i in range(0, len(points_in_polygon)-1):
        resto = datetime.datetime.strptime(points_in_polygon['timestamp'][i+1], "%d-%m-%Y %H:%M:%S")-datetime.datetime.strptime(points_in_polygon['timestamp'][i], "%d-%m-%Y %H:%M:%S")
        resto_minutes = resto.total_seconds()/60
        #si la diferencia entre los timestamp del omnibus en el putno supera los 40 minutos es q dio un viaje y regreso, sino lo supera es que se mantuvo ese tiempo esperando en el punto
        if (resto_minutes > 40):
            start_travel_vector.insert(len(start_travel_vector),i)
    start_travel_vector.insert(len(start_travel_vector),len(points_in_polygon)-1)
    return start_travel_vector
```

```python
#funcion que te calcula la distancia de todos los puntos dentro del poligono con respecto a un punto P
def new_column_distance_to_point(df_inside_polygon,point1,point2):
    from scipy.spatial import distance
    dist_vector = []
    dist_vector1 = []
    
    #creando una nueva columna al dataframe "latlon" que es la union de las columnas latitude y longitude
    df_inside_polygon['latlon'] = zip(df_inside_polygon.latitude, df_inside_polygon.longitude)
    
    for i in range(0,len(df_inside_polygon)):
        dist_vector.insert(len(dist_vector),distance.euclidean(df_inside_polygon['latlon'][i],point1))
        dist_vector1.insert(len(dist_vector1),distance.euclidean(df_inside_polygon['latlon'][i],point2))

    #si la columna 'distance' no existe, creala e inserta en ella los valores q están en el arreglo dist_vector
    if not 'distance_to_point1' in df_inside_polygon:
        newName = 'distance_to_point1'
        values = np.insert(df_inside_polygon.values,df_inside_polygon.shape[1],dist_vector,axis=1)
        header = df_inside_polygon.columns.values.tolist()
        header.append(newName)
        df_inside_polygon = pd.DataFrame(values,columns=header)
    
    if not 'distance_to_point2' in df_inside_polygon:
        newName = 'distance_to_point2'
        values = np.insert(df_inside_polygon.values,df_inside_polygon.shape[1],dist_vector1,axis=1)
        header = df_inside_polygon.columns.values.tolist()
        header.append(newName)
        df_inside_polygon = pd.DataFrame(values,columns=header)
    
    return df_inside_polygon
```

```python
#Funcion que recibe un dataframe con todas las observaciones de un viaje, el nombre del pto de inicio y nombre pto fin
#Retorna un dataframe con el tiempo de viaje 
def time_compute (df_travel, name_punto_inicio, name_punto_fin):
    nearest_ini = []
    nearest_fin = []
    time_table = []
    
    for i in range(0,len(df_travel)):
        #seleccionando todas las obeservaciones que están a menos de 200 metros de distancia del punto inicial 
        # 0.001903852= 200m, 0.00285617 = 300m
        if (df_travel['distance_to_point1'][i] < 0.00285617):
            nearest_ini.insert(len(nearest_ini),df_travel['distance_to_point1'][i])
            nearest_ini.sort()#se ordenan de menor a mayor distancia del punto2 
        
        #seleccionando todas las obeservaciones que están a menos de 200 metros de distancia del punto inicial 
        # 0.001903852= 200m
        if (df_travel['distance_to_point2'][i] < 0.00285617):
            nearest_fin.insert(len(nearest_fin),df_travel['distance_to_point2'][i])
            nearest_fin.sort()#se ordenan de menor a mayor distancia del punto2 
    
    #comprobando que haya al menos una observacion proxima al pto incial y una al punto final
    #es decir que los arreglos nearest_ini and nearest_fin no estén vacios
    if (nearest_ini and nearest_fin):
        #se guarda toda la tupla de menor distancia al punto1
        travel_start = df_travel[(df_travel['distance_to_point1'] == nearest_ini[0])]
        #recuperando el indice de la tupla   travel_start
        index1 = travel_start.index.values[0] 
        #hora que salio el omnibus del punto inicial del segmento
        initial_hour = travel_start ['timestamp'][index1]
        #convertir la hora de inicio de string a tipo datetime
        initial_hour = datetime.datetime.strptime(initial_hour, "%d-%m-%Y %H:%M:%S")

        #se guarda toda la tupla del de menor distancia al punto2
        travel_end = df_travel[(df_travel['distance_to_point2'] == nearest_fin[0])]
        #recuperando el indice de la tupla travel_end
        index2 = travel_end.index.values[0]  
        #hora que llegó el omnibus al punto final del segment
        final_hour = travel_end ['timestamp'][index2]
        #convertir la hora de fin de string a tipo datetime
        final_hour = datetime.datetime.strptime(final_hour, "%d-%m-%Y %H:%M:%S")
        
        #computando el tiempo de viaje
        travel_time = final_hour - initial_hour
        travel_time_in_minutes = travel_time.total_seconds()/60
        
        bus_line = travel_start ['line'][index1]
        bus_Id = travel_start ['bus_id'][index1]
        direction = name_punto_inicio + "_to_" + name_punto_fin
         
        time_table.insert(len(time_table), [bus_line, bus_Id, direction, initial_hour, final_hour, travel_time_in_minutes])
               
    
    #Construir el dataframe   
    mydf = pd.DataFrame(time_table, columns=['Line', 'Bus_Id','Segment', 'Arrive_Hour', 'Departure_Hour', 'Travel_Segment_Time'])       
    return mydf 
```

```python
#Funcion que crea la tabla de viajes con los tiempos de viaje
def segment_travel_table (name_A, arrive_array_A, name_B, departure_array_B, bus_Id, bus_line):
    
    my_time_table = []   
    num_travel = 0    
    i = 0 #Cursor para moverme por el departure_array_A(arreglo q contiene las horas de salida del punto incial)   
    
   
    while (i<len(departure_array_B)-1):
        if (arrive_array_A[i] <= departure_array_B[i]):
            direction = name_A + "_to_" + name_B
            departure_hour = departure_array_B[i]
            arrive_hour = arrive_array_A[i]
            #travel_time = bus.days_hours_minutes(departure_hour - arrive_hour) 
            #computando el tiempo de viaje
            travel_time = departure_hour - arrive_hour
            travel_time_in_minutes = travel_time.total_seconds()/60
                        
            i = i +1
            my_time_table.insert(len(my_time_table), [bus_line, bus_Id, direction, arrive_hour, departure_hour, travel_time_in_minutes])
                        
    
    #Construir el dataframe pasandole el array con todos los datos de los viajes   
    mydf = pd.DataFrame(my_time_table, columns=['Line', 'Bus_Id','Segment', 'Arrive_Hour', 'Departure_Hour', 'Travel_Segment_Time'])       
    return mydf 
```

```python
#lo estoy usando solo para viajes de un solo sentido
def bus_departure_hour(initial_points_timestamp_list):
    list_departure_hours = []
    i = 0
    if len(initial_points_timestamp_list) >1:
        for i in range(0, len(initial_points_timestamp_list)-1):
            resto = datetime.datetime.strptime(initial_points_timestamp_list[i+1], "%d-%m-%Y %H:%M:%S")-datetime.datetime.strptime(initial_points_timestamp_list[i], "%d-%m-%Y %H:%M:%S")
            resto_total = resto.total_seconds()/60
            #si la diferencia entre los timestamp del omnibus en el putno supera los 40 minutos es q dio un viaje y regreso, sino lo supera es que se mantuvo ese tiempo esperando en el punto
            if (resto_total > 40):
                 list_departure_hours.insert(len( list_departure_hours),datetime.datetime.strptime(initial_points_timestamp_list[i], "%d-%m-%Y %H:%M:%S"))
        #siempre es necesario insertar el ultimo valor de la lista de "initial_points_timestamp" en la lista de  list_departure_hours 
        list_departure_hours.insert(len( list_departure_hours),datetime.datetime.strptime(initial_points_timestamp_list[i+1], "%d-%m-%Y %H:%M:%S"))
    elif len(initial_points_timestamp_list) == 1:
        list_departure_hours.insert(len( list_departure_hours),datetime.datetime.strptime(initial_points_timestamp_list[0], "%d-%m-%Y %H:%M:%S"))
    else:
        list_departure_hours = []
    return list_departure_hours
```

```python
#lo estoy usando solo para viajes de un solo sentido
#Funcion que retorna el arreglo con las horas de llegada del omnibus al punto inicial/final del trayecto
def bus_arrive_hour(initial_points_timestamp_list):
    list_arrive_hours = []
    #siempre es necesario insertar el ultimo valor de la lista de "initial_points_timestamp" en la lista de leblon_departure_hours 
    if len(initial_points_timestamp_list) >1:
        list_arrive_hours.insert(len(list_arrive_hours),datetime.datetime.strptime(initial_points_timestamp_list[0], "%d-%m-%Y %H:%M:%S"))
        for i in range(0, len(initial_points_timestamp_list)-1):
            resto = datetime.datetime.strptime(initial_points_timestamp_list[i+1], "%d-%m-%Y %H:%M:%S")-datetime.datetime.strptime(initial_points_timestamp_list[i], "%d-%m-%Y %H:%M:%S")
            resto_total = resto.total_seconds()/60
            if (resto_total > 40):
                list_arrive_hours.insert(len(list_arrive_hours),datetime.datetime.strptime(initial_points_timestamp_list[i+1], "%d-%m-%Y %H:%M:%S"))
    elif (len(initial_points_timestamp_list) == 1):
        list_arrive_hours.insert(len(list_arrive_hours),datetime.datetime.strptime(initial_points_timestamp_list[0], "%d-%m-%Y %H:%M:%S"))
    else:
        list_arrive_hours = []        
    return list_arrive_hours  
```

```python
#Funcion que dada un string de connexion, un poligono y una lista de onmibus que pasan por aquel poligono
#Retorna el tiempo que demora cada bus_id en atravesar el tramo dentro del poligono
def for_one_way_street(conn_string, inverted_polygon, bus_line_list, punto_inicio, punto_fin):
    import bus
    new_df = pd.DataFrame(columns=['Line', 'Bus_Id','Segment', 'Arrive_Hour', 'Departure_Hour', 'Travel_Segment_Time']) 
   
    #Retorna todos las observaciones de todos las lineas que pasaron por el segmento aquel dia
    df_points_in_segment = points_inside_polygon(conn_string, inverted_polygon)
    
    #después que ya se determinaron los viajes en el sentido deseado
    for line in bus_line_list:
        #Recuperando las observaciones dentro del segmento por línea
        df_buslines_inside_polygon = df_points_in_segment[(df_points_in_segment['line']== line)]
        
        df_buslines_inside_polygon.groupby('line').bus_id.nunique()
        bus_id_list = pd.unique(df_buslines_inside_polygon.bus_id.ravel())
        myentry = []
        myout = []
        for bus_id in bus_id_list:
            df_busId_inside_polygon = df_buslines_inside_polygon[(df_buslines_inside_polygon['bus_id']== bus_id)]
            timestamp_busId_inside_polygon = df_busId_inside_polygon ['timestamp']
            timestamp_busId_inside_polygon.index = range(0,len(timestamp_busId_inside_polygon))
            #uso las funciones bus bus_arrive_hour y bus_departure_hour para delimitar los viajes
            myentry = bus.bus_arrive_hour(timestamp_busId_inside_polygon)
            myout = bus.bus_departure_hour(timestamp_busId_inside_polygon)
            result = segment_travel_table(punto_inicio, myentry, punto_fin, myout, bus_id, line)
            new_df = new_df.append(result, ignore_index = True)
            
    return new_df
    
```

```python
#Funcion que dado un fecha, un dataframe con todos las observaciones dentro de un poligono,
#lista de omnibus coorespondientes al tramo y un punto
#Retorna un dataframe con todos los viajes que van el sentido opuesto a ese punto
def for_two_way_streets(conn_string, inverted_polygon, bus_line_list, coord_initial_point,coord_final_point, punto_inicio, punto_fin):
    import bus
    new_df = pd.DataFrame(columns=['Line', 'Bus_Id','Segment', 'Arrive_Hour', 'Departure_Hour', 'Travel_Segment_Time'])
    df_timestamp_desired_direction = pd.DataFrame(columns=['timestamp'])
    
    #Retorna todos las observaciones de todos las lineas que pasaron por el segmento aquel dia
    df_points_in_segment = points_inside_polygon(conn_string, inverted_polygon)
    
    #después que ya se determinaron los viajes en el sentido deseado
    for line in bus_line_list:
        
        df_busId_inside_polygon_dist = []
        #Recuperando las observaciones dentro del segmento por línea
        df_buslines_inside_polygon = df_points_in_segment[(df_points_in_segment['line']== line)]
        #df_buslines_inside_polygon = df_points_in_segment[(df_points_in_segment['line']== '309')]
        
        df_buslines_inside_polygon.groupby('line').bus_id.nunique()
        bus_id_list = pd.unique(df_buslines_inside_polygon.bus_id.ravel())
                
        myinput= []
        myout = []
        for bus_id in bus_id_list:
            
            df_busId_inside_polygon = df_buslines_inside_polygon[(df_buslines_inside_polygon['bus_id']== bus_id)]
            travel_index_list = travels_indices(df_busId_inside_polygon)
            
            #adiciono la columna distancia al punto al dataframe df_busId_inside_polygon
            df_busId_inside_polygon_dist = new_column_distance_to_point(df_busId_inside_polygon,coord_initial_point, coord_final_point)
            
            #seleccionando solo los viajes que van en la direccion deseada (alejandose del punto inicial)
            
            #df_busId_C41383 = df_busId_inside_polygon_dist[(df_busId_inside_polygon_dist['bus_id']== 'C41383')]
            desired_direction_travels = travels_desired_direction(df_busId_inside_polygon_dist, travel_index_list,punto_inicio, punto_fin)

            #print desired_direction_travels
  
            new_df = new_df.append(desired_direction_travels, ignore_index = True)
            
                    
    return new_df
```

```python
#Funcion para separar la hora de la fecha en la columna Arrive_Hour
def convert_to_hour(column):
    from datetime import datetime
    list_time = []
    for row in column:
        str_row = str(row)
        split_time_array = str_row.split(' ')
        t_time = datetime.strptime(split_time_array[1], '%H:%M:%S').time()
        list_time.insert(len(list_time),t_time)
    return list_time
```

```python
#Funcion que recibe una dataframe con los travel time, una lista de intervalos
#Retorna un dataframe del promedio del tiempo de viaje por intervalo
def mean_travel_time(df_travel_time, day_hours_interval):
    
    segment = df_travel_time['Segment'][0]
    #convertir a formato datetime la columna 'Arrive_Hour'
    array_hours = convert_to_hour(df_travel_time['Arrive_Hour'])
    
    #crear una columna para almacenar el formato datetime    
    if not 'Arrive_Hour_datetime' in df_travel_time:
        newName = 'Arrive_Hour_datetime'
        values = np.insert(df_travel_time.values,df_travel_time.shape[1],array_hours,axis=1)
        header = df_travel_time.columns.values.tolist()
        header.append(newName)
        df_travel_time = pd.DataFrame(values,columns=header)
    
    #convertir el arreglo "day_hours_interval" a formato datetime
    temp_array = []
    for i in range(0,len(day_hours_interval)):
        temp_array.insert(len(temp_array),datetime.strptime(day_hours_interval[i], '%H:%M:%S').time()) 
    
    mean_travel_time = []
    count = 0
    for j in range(0,len(day_hours_interval)-1):
        df_temp = df_travel_time[(df_travel_time['Arrive_Hour_datetime']>= temp_array[j])& (df_travel_time['Arrive_Hour_datetime']< temp_array[j+1])]
        suma = 0
        for i in range(count, len(df_temp)+count):
            suma = suma + df_temp['Travel_Segment_Time'][i]
            count = count + 1
        if (len(df_temp)> 0):
            mean_time = suma/len(df_temp)
        else:  
            mean_time = 0
            
        time_interval = str(temp_array[j]) + ' - '+ str(temp_array[j+1])
        mean_travel_time.insert(len(mean_travel_time), [segment, time_interval, mean_time])
    
    df_mean_travel_time = pd.DataFrame(mean_travel_time,columns=['Segment','Time_Interval', 'Mean_Travel_Time'])
    return df_mean_travel_time
```

```python
"""Funciones para renderizar el mapa en el html"""
from IPython.display import HTML
 
def inline_map(map):
    """
    Embeds the HTML source of the map directly into the IPython notebook.
    
    This method will not work if the map depends on any files (json data). Also this uses
    the HTML5 srcdoc attribute, which may not be supported in all browsers.
    """
    map._build_map()
    return HTML('<iframe srcdoc="{srcdoc}" style="width: 100%; height: 510px; border: none"></iframe>'.format(srcdoc=map.HTML.replace('"', '&quot;')))
 
def embed_map(map, path="map.html"):
    """
    Embeds a linked iframe to the map into the IPython notebook.
    
    Note: this method will not capture the source of the map into the notebook.
    This method should work for all maps (as long as they use relative urls).
    """
    map.create_map(path=path)
    return HTML('<iframe src="files/{path}" style="width: 100%; height: 510px; border: none"></iframe>'.format(path=path))
```
