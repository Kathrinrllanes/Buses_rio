```python
import networkx as nx
import numpy as np
import pandas as pd
import json
import smopy
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = mpl.rcParams['savefig.dpi'] = 300

#Python libraries such as NumPy and SciPy
```

```python
#convertir de un csv a un shapefile (shp)
import shapefile as shp
import csv
out_file = '/home/kathrin/Doctorado/bus_rio/Codigos_Ipython/data/gtfs_linha317-shapes.shp'

#Set up blank lists for data
#linha,descricao,agencia,sequencia,shape_id,latitude,longitude

latitude,longitude,shape_id = [],[],[]

#read data from csv file and store in lists
with open('/home/kathrin/Doctorado/bus_rio/Datos_Omnibus/Onibus/Shapes/gtfs_linha317-shapes.csv', 'rb') as csvfile:
    r = csv.reader(csvfile, delimiter=',')
    for i,row in enumerate(r):
        if i > 0: #skip header
            latitude.append(float(row[5]))
            longitude.append(float(row[6]))
            #linha.append(float(row[2])
            #descricao.append(row[3])
            #agencia.append(row[4])
            #sequencia.append(row[5])
            shape_id.append(float(row[4]))
                            
result = shp.Writer(shp.POINT)
result.autoBalance = 1 #ensures gemoetry and attributes match
result.field('latitude','F',10,8)
result.field('longitude','F',10,8)
#w.field('linha','D')
#w.field('descricao','C')
#w.field('agencia','A')
#w.field('sequencia','S')
result.field('shape_id','N')
                            

#loop through the data and write the shapefile
for j,k in enumerate(latitude):
    result.point(k,longitude[j]) #write the geometry
    result.record(k,longitude[j],shape_id[j]) #write the attributes
 
#Save shapefile
result.save(out_file)
```

```python
#funcion que retorna un grafo que conecta todas las posiciones almacenadas ahi
#g = nx.read_shp("/home/kathrin/Doctorado/bus_rio/Codigos_Ipython/data/data/tl_2013_06_prisecroads.shp")
g = nx.read_shp("/home/kathrin/Doctorado/bus_rio/Codigos_Ipython/data/gtfs_linha317-shapes.shp")
```

```python
#creando un grafo conectado 
sg = list(nx.connected_component_subgraphs(g.to_undirected()))[0]
len(sg)
```

```python
#pos0 = (36.6026, -121.9026)
#pos1 = (34.0569, -118.2427)
pos0 = (-22.9049,-43.1917)
pos1 = (-23.0012, -43.3649)
```

```python
#funcion q retorna un arreglo de coordenadas contenidas entre dos puntos n0 y n1 a trav√©s de una road
def get_path(n0, n1):
    """If n0 and n1 are connected nodes in the graph, this function
    return an array of point coordinates along the road linking
    these two nodes."""
    return np.array(json.loads(sg[n0][n1]['Json'])['coordinates'])
```

```python
#funcion para calcular la distancia entre 2 puntos teniendo en cuenta el radio de la tierra
EARTH_R = 6372.8
def geocalc(lat0, lon0, lat1, lon1):
    """Return the distance (in km) between two points in 
    geographical coordinates."""
    lat0 = np.radians(lat0)
    lon0 = np.radians(lon0)
    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    dlon = lon0 - lon1
    y = np.sqrt(
        (np.cos(lat1) * np.sin(dlon)) ** 2
         + (np.cos(lat0) * np.sin(lat1) 
         - np.sin(lat0) * np.cos(lat1) * np.cos(dlon)) ** 2)
    x = np.sin(lat0) * np.sin(lat1) + \
        np.cos(lat0) * np.cos(lat1) * np.cos(dlon)
    c = np.arctan2(y, x)
    return EARTH_R * c
```

```python
#funcion que calcula el largo del camino entre un nodo n1 y un nodo n2
def get_path_length(path):
    return np.sum(geocalc(path[1:,0], path[1:,1],
                          path[:-1,0], path[:-1,1]))
```

```python
# Compute the length of the road segments.
for n0, n1 in sg.edges_iter():
    path = get_path(n0, n1)
    distance = get_path_length(path)
    sg.edge[n0][n1]['distance'] = distance
```

```python
nodes = np.array(sg.nodes())
# Get the closest nodes in the graph.
pos0_i = np.argmin(np.sum((nodes[:,::-1] - pos0)**2, axis=1))
pos1_i = np.argmin(np.sum((nodes[:,::-1] - pos1)**2, axis=1))
```

```python
# Compute the shortest path.
path = nx.shortest_path(sg, 
                        source=tuple(nodes[pos0_i]), 
                        target=tuple(nodes[pos1_i]),
                        weight='distance')
len(path)
```

```python
# list of edges that form the shortest path between our two positions
roads = pd.DataFrame([sg.edge[path[i]][path[i + 1]] 
                      for i in range(len(path) - 1)], 
                     columns=['FULLNAME', 'MTFCC', 
                              'RTTYP', 'distance'])
roads
```

```python
#total length of this itinerary
roads['distance'].sum()
```

```python
#display the itinerary on the map
map = smopy.Map(pos0, pos1, z=7, margin=.1)
```

```python
#function that concatenates the positions along every edge in the path
def get_full_path(path):
    """Return the positions along a path."""
    p_list = []
    curp = None
    for i in range(len(path)-1):
        p = get_path(path[i], path[i+1])
        if curp is None:
            curp = p
        if np.sum((p[0]-curp)**2) > np.sum((p[-1]-curp)**2):
            p = p[::-1,:]
        p_list.append(p)
        curp = p[-1]
    return np.vstack(p_list)
```

```python
#convert the path in pixels in order to display it on the Smopy map.
linepath = get_full_path(path)
x, y = map.to_pixels(linepath[:,1], linepath[:,0])
```

```python
#display the map, with our two positions and the computed itinerary between them.
plt.figure(figsize=(6,6));
map.show_mpl();
# Plot the itinerary.
plt.plot(x, y, '-k', lw=1.5);
# Mark our two positions.
plt.plot(x[0], y[0], 'ob', ms=10);
plt.plot(x[-1], y[-1], 'or', ms=10);
```
